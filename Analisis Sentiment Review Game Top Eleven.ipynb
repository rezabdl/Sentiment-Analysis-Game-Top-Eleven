# Text Processing

*Cleaning text* : pada bagian ini dilakukan pembersihan text dari simbol-simbol, emoticon yang nantinya akan menghambat proses-proses selanjutnya 

```python
# Membuat sebuah variabel baru hanya untuk mempermudah proses saja 
data_kedua = data_pertama[['content','Label']]
data_kedua
```

```python
# Mengecek apakah isi pada atribut content atau pada data_kedua apakah terdapat missing value atau tidak
data_kedua.isna().sum()
```

```python
# Membuat fungsi untuk memproses kalimat, jika terdapat simbol-simbol unik dapat dihapuskan
def cleaning_data(data_kedua,text_field,new_text_field_name) :
    data_kedua[new_text_field_name] : data_kedua[text_field].str.lower()
    data_kedua[new_text_field_name] = data_kedua[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))  
    # remove numbers
    data_kedua[new_text_field_name] = data_kedua[new_text_field_name].apply(lambda elem: re.sub(r"\d+", "", elem)) 
    return data_kedua

def praproses(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=)()(?:-)?(?:\)|\(|D|P)',
                           text)
    text = (re.sub('[\W]+', ' ', text.lower()) +
            ' '.join(emoticons).replace('-', ''))
    return text

def hapus_katadouble(s): 
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    return pattern.sub(r"\1\1", s)

def clean_text_4(text):
    text = ' '.join(word for word in str(text).split() if len(word) > 3)
    return text

```

```python
# Membuat variabel baru yang berisi kalimat yang sudah dibersihkan dan sudah sama besarnya
data_kedua['text'] = data_kedua['content'].str.lower().apply(praproses).apply(hapus_katadouble).apply(clean_text_4)
data_bersih = cleaning_data(data_kedua,'content','text')
data_bersih.head(10)
```

```python
# # Data yang sudah bersih di atas, saya eksport ke csv untuk memudahkan kegiatan selanjutnya
data_kedua.to_csv('Data_Sudah_Cleaning.csv', index=False)
```

*Filtering* : adalah proses untuk menyeleksi kata kata yang tidak memiliki makna maka akan dihapuskan

```python
# disini kita import library yang digunakan untuk melakukan proses ini 
import nltk.corpus
from nltk.corpus import stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# disini kita akan melakukan stopword atau menghapus kata kata yang tidak bermakna akan dihapus membutuhkan stopwords.words('indonesian') dikarenakan yang akan kita eksekusi adalah kata kata dalam bahasa indonesia
stop = stopwords.words('indonesian')
data_bersih['text_StopWord'] = data_bersih['text'].apply(lambda x:' '.join([word for word in x.split() if word not in (stop)]))


sastrawi = StopWordRemoverFactory()
stopword = sastrawi.get_stop_words()

def stop_word(text) :
    text = ' '.join(word for word in text.split() if word not in stopword)
    return text

```

```python
data_bersih['text_stopword'] = data_bersih['text_StopWord'].apply(stop_word)
data_bersih
```

*Tokenizing* text : proses ini adalah untuk memisahkan teks menjadi unit-unit kecil yang disebut "token". Token bisa berupa kata, frasa, atau karakter, tergantung pada tingkat kompleksitas yang diinginkan.

```python
# Import library yang digunakan untuk proses tokenizing 
import nltk
from nltk.tokenize import word_tokenize
```

```python
# disini digunakan untuk melakukan tokenizing
data_bersih['text_tokens'] = data_bersih['text_stopword'].apply(lambda x: word_tokenize(x))
data_bersih.head()
```

# Proses Stemming


```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
#-----------------STEMMING -----------------
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
#import swifter


# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}
hitung=0

for document in data_bersih['text_tokens']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")
for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    hitung+=1
    print(hitung,":",term,":" ,term_dict[term])

print(term_dict)
print("------------------------")

# mengaplikasikan stemming ke data_bersihframe 
def get_stemmed_term(document):
    return [term_dict[term] for term in document]


```

```python
#script ini bisa dipisah dari eksekusinya setelah pembacaaan term selesai
data_bersih['text_steamindo'] = data_bersih['text_tokens'].apply(lambda x:' '.join(get_stemmed_term(x)))
data_bersih.head(20)
```
